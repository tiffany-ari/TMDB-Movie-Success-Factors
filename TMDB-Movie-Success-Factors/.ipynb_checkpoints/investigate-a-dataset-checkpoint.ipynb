###### From getting a broad overview of the dataset, I was able to find out:
* There are 10866 rows, 21 columns.
* The release_date data type is the only one that would need to be converted, if used.
* There are many null values, especially in the homepage, tagline, keywords, and production_companies columns.
* There are many zero-values in the budget, revenue, budget_adj, and reevenue_adj columns, which may be missing data.

## Cleaning process:

##### 1. Drop columns
* I will be dropping any columns that I won't be needing
* I will also be dropping the original budget and revenue columns, and instead be using the budget_adj and revenue_adj since they have been adjusted for inflation and so will be more relevent. 

```python
# Drop columns
df.drop(['imdb_id', 'budget', 'revenue', 'homepage', 'tagline', 'keywords', 'overview', 
         'production_companies', 'release_date', ], axis=1, inplace=True)
```

```python
# Check if columns were dropped
df.head()
```

##### 2. Check for duplicates
* Drop duplicates, if any.

```python
# Check for duplicates
sum(df.duplicated())
```

```python
# drop duplicates
df.drop_duplicates(inplace=True)
```

##### 3. Split cast names
* Since some cast names are combined, I will split them into their individual names for better analysis.

```python
df['cast'].value_counts()
```

```python
df['cast'] = df['cast'].str.split("|")
df.head(1)
```

```python
cast_df = df.explode('cast')
cast_df.head(1)
```

```python
cast_df['cast'].value_counts()
```

##### 4. Handle missing data
* I will drop the null values in the director, cast, and genre columns, since there are not that many. 
* There are a significant amount of zero-data entries in the budget_adj and revenue_adj. These are probably missing data, but as they make up almost half of the values in those columns, I will be replacing them with NaN to preserve the data integrety.

```python
# Check for null values
df.isnull().sum()
```

```python
# Drop null values
df.dropna(subset=['director', 'cast', 'genres'], inplace=True)
```

```python
# Confirm values were dropped
df.isnull().sum()
```

```python
# replace missing values with null
df['budget_adj'] = df['budget_adj'] .replace(0, np.nan)
df['revenue_adj'] = df['revenue_adj'].replace(0, np.nan)
```

```python
# confirm missing values in adjusted budget and revenue are replaced
df.describe()
```

##### 5. Add new column
* I will be adding a 'decades' column, to see if there are any revenue trends between decades.

```python
# overview of release years, to make decades column
df['release_year'].describe()
```

```python
# bin edges that will be used to "cut" the data into groups
bin_edges = [1960 , 1970, 1980, 1990, 2000, 2010, 2016]
```

```python
# Labels for the decade groups
bin_names = ['1960', '1970', '1980', '1990', '2000', '2010']
```

```python
# Create decades column
df['decades'] = pd.cut(df['release_year'], bin_edges, labels=bin_names, include_lowest=True)
df.head(1)
```

##### 6. Convert currency
* I am converting dollars into millions of dollars for easier graph visualization.

```python
# convert dollars into millions
df['budget_adj ($Million)'] = df['budget_adj'].div(10**6)
df['revenue_adj ($Million)'] = df['revenue_adj'].div(10**6)
```

## Exploratory Data Analysis



### Are movies more popular if they have a higher budget?

```python
# Histogram of adjusted budget
df['budget_adj ($Million)'].hist(figsize=(8,5))
plt.title('Adjusted Budget')
plt.xlabel('Budget')
plt.ylabel('Number of Movies');
```

```python
# Histogram of popularity
df['popularity'].hist(figsize=(8,5))
plt.title('Popularity')
plt.xlabel('Popularity Score')
plt.ylabel('Number of Movies');
```

I made two histograms of budget and popularity, to give an overview. They look fairly similar by themselves, so I will compare them in another graph to get a better understading of the relationship between the two variables.

```python
# function for scatter plots
def scatterPlots(df, var1, var2):
    sns.lmplot(x=var1, y=var2, data=df)  # plot graph
    
    # clean up label names
    var1=var1.replace('_', ' ')
    var2=var2.replace('_', ' ')
    
    # add title & x, y labels 
    plt.title(f'{var1} vs. {var2}'.title(), fontsize=12)
    plt.xlabel(var1.title())
    plt.ylabel(var2.title())
    
```

```python
# call scatter plot function
scatterPlots(df,'budget_adj ($Million)','popularity')
```

From this graph, it looks like having a higher budget may have a positive impact on popularity scores. However, the outliers seem to show that having a very high budget may not have a positive impact on popularity, as the most popular movies are in the mid-budget range. 

```python
# call scatter plot function
scatterPlots(df,'budget_adj ($Million)', 'vote_average')
```

I also plotted adjusted budget against the average viewer votes, to compare and see the difference. From this graph, it doesn't look like having a higher budget makes much of an impact on vote averages. The vote average actually looks like its higher when the budget is lower.

### Does a higher budget mean a higher revenue?

```python
# call scatter plot function
scatterPlots(df, 'budget_adj ($Million)', 'revenue_adj ($Million)')
```

This first scatterplot is showing the relationship between budget and revenue. It looks like a higher budget is associated with movies having a higher revenue as well. However, too high of a budget looks like it may result in losses. 

```python
mean_budget = df['budget_adj ($Million)'] .mean()
mean_budget
```

```python
mean_revenue = df['revenue_adj ($Million)'] .mean()
mean_revenue
```

I want to compare the mean budget and revenue together, to see if maybe it can give more insight into the variables, and their relationship.

```python
locations = [1, 2]
heights = [mean_budget, mean_revenue]
labels = ['Budget', 'Revenue']
plt.bar(locations, heights, tick_label=labels)
plt.title('Mean Budget vs. Revenue (Adjusted)')
plt.ylabel('$ Hundred Million');
```

The mean budget is almost 300 million, while the revenue is about 800 million. From the scatterplot it looks like the majority of movies below the mean make a higher revenue than movies above the mean budget, which is interesting. However, there are a lot of outliers in this dataset.

### Correlation Plots
These are some plots to further investigate any possible correlations between categories that I think are the most relevant to a films success.

```python
#information on top revenue films
movie_info = pd.DataFrame(df['revenue_adj ($Million)'].sort_values(ascending=False))
data_set = ['popularity','budget_adj ($Million)','release_year','original_title','cast','director','genres', 'runtime','vote_average','release_year']
for i in data_set:
    movie_info[i] = df[i]
movie_info.head(5)
```

```python
def plot_correlation(df):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 10 , 8 ) )
    cmap = sns.diverging_palette( 240 , 10 , as_cmap = True )
    _ = sns.heatmap(corr,cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 })

```

```python
#heatmap plot
plot_correlation(df[['popularity','budget_adj ($Million)','release_year','revenue_adj ($Million)','runtime','vote_average', 'popularity']])
```

```python
# plot of revenue vs. buget, runtime, popularity, vote average
fig, axes = plt.subplots(2,2,figsize = (20,10))
fig.suptitle("Revenue Vs (Budget, Runtime, Popularity, & Vote Average)",fontsize=14)
sns.regplot(x=df['revenue_adj ($Million)'], y=df['budget_adj ($Million)'],color='c',ax=axes[0][0])
sns.regplot(x=df['revenue_adj ($Million)'], y=df['runtime'],color='c',ax=axes[0][1])
sns.regplot(x=df['revenue_adj ($Million)'], y=df['vote_average'],color='c',ax=axes[1][0])
sns.regplot(x=df['revenue_adj ($Million)'], y=df['popularity'],color='c',ax=axes[1][1])

sns.set_style("whitegrid")
```

***Revenue vs Budget:*** The above plot shows that films with higher budgets have slightly higher revenues. There are few movies with very high budgets however, but there seems to be a moderate positive correlation with a coefficent value of 0.57.

***Revenue vs Runtime:*** Revenue and runtime have a correlation coefficient of 0.26. This, along with the above plot, shows that runtime is not a significant factor in revenue.

***Revenue vs vote average:*** The above graph shows a weak positive correlation revenue and vote average, and as the correlation coefficien is 0.24, there is not much influence between vote average and movie revenue.

***Revenue vs Popularity:*** According to the above plot and heatmap, popularity does seem to have an influence on revenue as there is a moderate positive correlation and a coefficient of 0.56.
